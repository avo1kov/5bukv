```python
import re
```

```python
with open('russian_nouns_5_letters.txt', 'r', encoding='utf-8') as file:
    nouns = [line.strip() for line in file]
```

```python
nouns = [noun.lower() for noun in nouns]
nouns = [noun.replace('ё', 'е') for noun in nouns]
nouns = [noun for noun in nouns if re.match(r'^[a-zA-Zа-яА-Я0-9\s]+$', noun)]
```

```python
# i need to calc frequency of each letter in the list of nouns. but every letter should be counted only once in each word
from collections import Counter

letter_frequency = Counter()

for noun in nouns:
    letters = set(noun)
    letter_frequency.update(letters)

print(letter_frequency)

# create a list of tuples with letter and its frequency
letter_frequency_list = list(letter_frequency.items())
print(letter_frequency_list)
```

```python
sum_of_frequencies = {}

for noun in nouns:
        sum_of_frequencies[noun] = sum(letter_frequency[letter] for letter in noun)

print(sum_of_frequencies)
```

```python
sorted_nouns = sorted(nouns, key=lambda x: (len(set(x)), sum_of_frequencies[x]), reverse=True)
```

```python
with open('sorted_nouns.txt', 'w') as file:
    for noun in sorted_nouns:
        file.write(noun + '\n')
```
